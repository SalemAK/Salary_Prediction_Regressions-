## Cleaning Data

```python
df.isnull().sum()
```

```python
clean_df = df
clean_df["AGE"] = clean_df["AGE"].fillna(clean_df["AGE"].median())
clean_df["DOJ"] = clean_df["DOJ"].fillna(clean_df["DOJ"].mode()[0])
clean_df["LEAVES USED"] = clean_df["LEAVES USED"].fillna(clean_df["LEAVES USED"].mean())
clean_df["LEAVES REMAINING"] = clean_df["LEAVES REMAINING"].fillna(clean_df["LEAVES REMAINING"].mean())
clean_df["RATINGS"]= clean_df["RATINGS"].fillna(clean_df["RATINGS"].mean())
clean_df = clean_df.drop_duplicates()
```

```python
# Check Number Nulls in each columns
clean_df.isnull().sum()
```

## Feature Engineering

```python
preprocessing_df = clean_df
preprocessing_df["DOJ"] = pd.to_datetime(preprocessing_df["DOJ"])
preprocessing_df["CURRENT DATE"] = pd.to_datetime(preprocessing_df["CURRENT DATE"])
preprocessing_df["Join_Years"] = (preprocessing_df["CURRENT DATE"] - preprocessing_df["DOJ"]).dt.days / 365.25
preprocessing_df["Total EXP"] = np.log2(preprocessing_df["PAST EXP"] + preprocessing_df["Join_Years"])
```

```python
print(preprocessing_df.dtypes)
```

## Data Preprocessing:

```python
preprocessing_df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, MinMaxScaler
```

```python
numerical_futures = preprocessing_df.select_dtypes(include=["float64", "int64"]).columns
numerical_futures = numerical_futures.drop("SALARY")
numerical_futures
```

```python
categorical_features
```

```python


encoder = OneHotEncoder(sparse_output=False)
encoded_data = encoder.fit_transform(preprocessing_df[categorical_features])
encoded_data = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_features))
preprocessing_df = pd.concat([preprocessing_df.drop(columns=categorical_features, axis=1).reset_index(drop=True), encoded_data], axis=1)
```

```python
preprocessing_df = preprocessing_df.drop(["DOJ", "CURRENT DATE"], axis=1)
```

#Split the data for train and validation then scale

```python
from sklearn.model_selection import train_test_split
X = preprocessing_df.drop(["SALARY"], axis=1)
y = preprocessing_df["SALARY"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2024)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train[numerical_futures])

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[numerical_futures] = scaler.transform(X_train[numerical_futures])
X_test_scaled[numerical_futures] = scaler.transform(X_test[numerical_futures])
```

#Model and Training
Algorithm

LR (Linear Regression)

PL(PolynomialFeatures)

SVR (Support Vector Regression)

DTR (DecisionTreeRegressor)

RFR (Random Forest Regression)

LGB (lightgbm)

CATB (catboost)

XGB(xgboost)


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

lr_model.fit(X_train_scaled, y_train)
y_pred = lr_model.predict(X_test_scaled)


lr_mae = mean_absolute_error(y_test, y_pred)
lr_mse = mean_squared_error(y_test, y_pred)
lr_rmse = np.sqrt(lr_mse)
lr_r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {lr_mae}")
print(f"Mean Squared Error: {lr_mse}")
print(f"Root Mean Squared Error: {lr_rmse}")
print(f"R-squared: {lr_r2}")
print("---------")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())

poly_model.fit(X_train_scaled, y_train)
y_pred = poly_model.predict(X_test_scaled)

poly_mae = mean_absolute_error(y_test, y_pred)
poly_mse = mean_squared_error(y_test, y_pred)
poly_rmse = np.sqrt(poly_mse)
poly_r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {poly_mae}")
print(f"Mean Squared Error: {poly_mse}")
print(f"Root Mean Squared Error: {poly_rmse}")
print(f"R-squared: {poly_r2}")
print("---------")
```

```python
from sklearn.svm import SVR

svr_model = SVR(kernel="linear")

svr_model.fit(X_train_scaled, y_train)
y_pred = svr_model.predict(X_test_scaled)

svr_mae = mean_absolute_error(y_test, y_pred)
svr_mse = mean_squared_error(y_test, y_pred)
svr_rmse = np.sqrt(svr_mse)
svr_r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {svr_mae}")
print(f"Mean Squared Error: {svr_mse}")
print(f"Root Mean Squared Error: {svr_rmse}")
print(f"R-squared: {svr_r2}")
print("---------")
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr_model = DecisionTreeRegressor()

dtr_model.fit(X_train_scaled, y_train)
y_pred = dtr_model.predict(X_test_scaled)

dtr_mae = mean_absolute_error(y_test, y_pred)
dtr_mse = mean_squared_error(y_test, y_pred)
dtr_rmse = np.sqrt(dtr_mse)
dtr_r2 = r2_score(y_test, y_pred)

# Print the metrics
print(f"Mean Absolute Error: {dtr_mae}")
print(f"Mean Squared Error: {dtr_mse}")
print(f"Root Mean Squared Error: {dtr_rmse}")
print(f"R-squared: {dtr_r2}")
print("---------")
```

#Tuning The best Model


*   Linear Regression




```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline


lr_model =  Pipeline([('regressor', LinearRegression())])

param_grid_lr = {
    'regressor__fit_intercept': [True, False],
    'regressor__copy_X': [True, False],
    'regressor__n_jobs': [-1, 1, 2, 4],
    'regressor__positive': [True, False]

}

grid_search_lr = GridSearchCV(lr_model, param_grid_lr, cv=5, scoring='r2')


grid_search_lr.fit(X_train_scaled, y_train)

print("Best parameters for Linear Regression:", grid_search_lr.best_params_)
print("Best R2 score for Linear Regression:", grid_search_lr.best_score_)
```

```python
import joblib
joblib.dump(grid_search_lr.best_estimator_, 'best_lr_model.pkl')
```
